# DUMMIES SYMPTOMS

```python
import pandas as pd

df_clean=pd.read_csv("2_clean_dataset.csv")
```

```python
s = []

for i in range(len(df_clean['SYMPTOM1'])):
    for k in range(10):
        column_name="SYMPTOM"+str(k+1)
        s.append(df_clean[column_name][i])
        
s = [x for x in s if type(x)==str]
```

```python
from collections import defaultdict
import collections

appearances = defaultdict(int)

for curr in s:
    appearances[curr] += 1
    
sorted_dict =(appearances)
sorted_dict = collections.OrderedDict(appearances)

sorted_x = sorted(sorted_dict.items(), key=lambda kv: kv[1])
sorted_x.reverse()
sorted_dict = collections.OrderedDict(sorted_x)
```

```python
top_symp=list(sorted_dict.keys())[:250]

for symp in top_symp:
    df_clean[symp]=0
    
for k in range(10):
    
    c_name="SYMPTOM"+str(k+1)
    print(c_name)
    
    for i in range(len(df_clean[c_name])):
        
        symp=df_clean[c_name][i]
        
        if( symp in top_symp):
            df_clean.loc[i,symp]=int("1")
```

```python
count=[]

for i in range(len(df_clean['Headache'])):
    x=0
    
    for c in df_clean.columns[48:]:
        x=x+int(df_clean[c][i])
        
   # print(x)
    count.append(x)
```

```python
df_clean['COUNT_SYMP']=count
```

```python
df=pd.read_csv("8_1_vaers_k_means_cluster_all_columns.csv")
df_s=df[df.columns[14:265]]

#seleziono il modello
df_s['Cluster_Label_5']=df['Cluster_Label_5'].tolist()
df_grouped=df_s.groupby(['Cluster_Label_5']).sum()

l_shapes=[]

for i in range(5):
    l_shapes.append(len(df[df['Cluster_Label_5']==i]))
```

```python
for c in df_grouped.columns:
    for i in range(5):
        df_grouped[c]=df_grouped[c].astype(float)
        val=round((df_grouped[c][i]/l_shapes[i])*100,2)
        df_grouped[c][i]=val
        
#df_grouped.to_csv("table.csv")

df_grouped
```
