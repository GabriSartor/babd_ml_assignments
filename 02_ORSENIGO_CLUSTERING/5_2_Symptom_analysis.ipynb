# SYMPTOMS ANALYSIS

```python
df=pd.read_csv("4_1_vaers_k_means_clusters_scaling.csv")
df_s=df[df.columns[14:265]]

#seleziono il modello
df_s['Cluster_Label_5']=df['Cluster_Label_5'].tolist()
df_grouped=df_s.groupby(['Cluster_Label_5']).sum()

l_shapes=[]

for i in range(5):
    l_shapes.append(len(df[df['Cluster_Label_5']==i]))
```

```python
for c in df_grouped.columns:
    for i in range(5):
        df_grouped[c]=df_grouped[c].astype(float)
        val=round((df_grouped[c][i]/l_shapes[i])*100,2)
        df_grouped[c][i]=val
        
#df_grouped.to_csv("table.csv")

df_grouped
```

```python
df_grouped.drop("COUNT_SYMP",axis=1,inplace=True)
```
